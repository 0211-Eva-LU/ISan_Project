```python
import requests
import json
import os
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
# from webdriver_manager.chrome import ChromeDriverManager

# 建立 Service 物件，指定 chromedriver.exe 的路徑


# 設定 Chrome 瀏覽器的選項
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized") # Chrome 瀏覽器在啟動時最大化視窗
options.add_argument("--incognito") # 無痕模式
options.add_argument("--disable-popup-blocking") # 停用 Chrome 的彈窗阻擋功能。

# 建立 Chrome 瀏覽器物件
driver = webdriver.Chrome(options=options)


folder_name = "projectimg"
if not os.path.exists(folder_name):
    os.mkdir('projectimg')
else:
    print("已建立過資料夾")


with open('movedetail2.json','r',encoding='utf-8') as f:
    data=json.load(f)

move_data=[]
for index,url_element in enumerate(data):
    # print(sticker_element)
    # print(sticker_element["data-preview"])
    url_element = data[index]["move_img"]
    move_id = data[index]["move_id"]
 
    move_data.append({"url":url_element,
                      "id":move_id})
#找到圖片
```

```python
'''
課堂練習
自行選擇三種貼圖進行下載
在儲存時要自動產生該貼圖名稱的資料夾並將圖片儲存在內
'''

def linestatic(number):

    import requests
    from bs4 import BeautifulSoup as bs
    import os

    url =f'https://store.line.me/stickershop/product/{number}/zh-Hant'
    r= requests.get(url)
    soup = bs(r.text,'lxml')


    sticker_name = soup.find("p",class_="mdCMN38Item01Ttl").text

    #建立資料夾
    # folder_name = sticker_name
    if not os.path.exists(sticker_name):
        os.mkdir(sticker_name)
    else:
        print("已建立過資料夾")



    #找到貼圖清單
    sticker = soup.find('ul',class_="mdCMN09Ul FnStickerList")
    #找到所有貼圖元素
    sticker_elements = sticker.find_all("li")



    for index,sticker_element in enumerate(sticker_elements):
        # print(sticker_element)
        # print(sticker_element["data-preview"])
        data_preview = json.loads(sticker_element["data-preview"])
        static_url = data_preview["staticUrl"]
        print(static_url)

        img_content = requests.get(static_url).content
        with open(f"{sticker_name}/line_sticker{sticker_name}_{index}.png","wb") as f:
            f.write(img_content)

linestatic(27469)
```
